### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = df['mpg'].median()
above_median_mpg_cars = df[df['mpg'] > median_mpg].copy()
above_median_mpg_cars['hp_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']
max_hp_to_weight_ratio = above_median_mpg_cars['hp_to_weight_ratio'].max()
best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight_ratio'] == max_hp_to_weight_ratio]

print("Car(s) with the best horsepower to weight ratio among those with above-median mpg:")
print(best_ratio_cars[['name', 'hp_to_weight_ratio', 'mpg']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
average_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(data=average_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')


plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
car_stats = df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
)

consistent_cars = car_stats[
    (car_stats['appearances'] >= 2) &
    (car_stats['std_mpg'] < 1.0)
]


min_std_mpg = consistent_cars['std_mpg'].min()
lowest_variation_cars = consistent_cars[consistent_cars['std_mpg'] == min_std_mpg]


result = lowest_variation_cars.reset_index()[['name', 'appearances', 'avg_mpg']]
result.columns = ['Model Name', 'Number of Appearances', 'Average MPG']

result_sorted = result.sort_values(by=['Number of Appearances', 'Average MPG'], ascending=[False, False])

print("Consistent car model(s) with the lowest variation in mpg:")
print(result_sorted)

```
